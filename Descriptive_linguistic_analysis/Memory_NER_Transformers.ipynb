# Named Entity recognition

```python
# Using spacy.load().
nlp = spacy.load("en_core_web_trf", disable = ["tagger", "parser", "attribute_ruler", "lemmatizer"])
```

Split the dataset in 5 parts to accelerate the process and to not run out of memory: 

1/5

```python
docs= []

for doc in nlp.pipe(df.loc[:1000,'story'], n_process=os.cpu_count()-1, batch_size = 1000):
    docs.append(doc)
```

```python
# Create an empty list to store the Entities 
ner = []

# Loop through each sentence in the dataset
for doc in docs:
    # Create a set of tuples to capture both the text (entity) and type of named entity (label):
    entities = set((entity.text, entity.label_) for entity in doc.ents)
    entity_counts = {}
    # Loop through each tuple in the document
    for entity, label in entities:
        # Check whether the current label has already been encountered. 
        # If not, create a new dictionary for that label with the current entity as a key and a count of 1. 
        if label not in entity_counts:
            entity_counts[label] = {entity: 1}
        # If the label is already in the dictionary but the current entity has not been encountered
        # we add a new key-value pair with a count of 1.
        elif entity not in entity_counts[label]:
            entity_counts[label][entity] = 1
        # If both the label and the entity are already in the dictionary
        # we increment the count for that entity.
        else:
            entity_counts[label][entity] += 1
    ner.append(entity_counts)

# Create a DataFrame from the dictionary with the frequency of each unique Entity for each Intention
ner_df = pd.DataFrame.from_records([{col: sum(inner_dict.values()) for col, inner_dict in d.items()} for d in ner]).fillna(0)
ner_df = ner_df.div(list(df.loc[:1000,'word_counts']), axis = 0)

# Display the resulting DataFrame
ner_df['word_counts'] = list(df.loc[:1000,'word_counts'])
ner_df['memType'] = list(df.loc[:1000,'memType'])
ner_df.head()
```

# Data Analysis

```python
# Load and append all the datasets.

# Load:
ner_1 = pd.read_csv('ner_hippocorpus_1000.csv')
ner_2 = pd.read_csv('ner_hippocorpus_2000.csv')
ner_3 = pd.read_csv('ner_hippocorpus_3000.csv')
ner_4 = pd.read_csv('ner_hippocorpus_4000.csv')
ner_5 = pd.read_csv('ner_hippocorpus_5000.csv')

# Concatenate all the datasets:
ner_df = pd.concat([ner_1, ner_2, ner_3, ner_4, ner_5]).fillna(0)
ner_df = ner_df.reset_index(drop=True) # stop resetting the row index

# Display the merged dataset
ner_df
```

```python
print("Entity\t\tDESCRIPTION")
print("---------------------------")

for ent in list(ner_df.columns[:-2]):
  print(ent+"\t\t"+spacy.explain(ent))
```

Group the columns representing the same construct: 

**Temporal Details**: Data + Time + Event

**Spatial Details**: GPE + Lol + Faculties

**People**: Person

**Numerical**: Percent + Money + Quantity + Cardinal + Ordinal

**Other Details (concrete reference to smth)**: Org + Norp + Language + Work_of_art + Product + Law 



```python
new_ner = pd.DataFrame(ner_df[['word_counts','memType']])
new_ner['Temporal'] = ner_df['DATE'] + ner_df['TIME'] + ner_df['EVENT']
new_ner['Spatial'] = ner_df['GPE'] + ner_df['LOC'] + ner_df['FAC']
new_ner['People'] = ner_df['PERSON']
new_ner['Numerical'] = ner_df['PERCENT'] + ner_df['MONEY'] + ner_df['QUANTITY'] + ner_df['CARDINAL'] + ner_df['ORDINAL']
new_ner['Others'] = ner_df['ORG'] + ner_df['NORP'] + ner_df['LANGUAGE'] + ner_df['WORK_OF_ART'] + ner_df['PRODUCT'] + ner_df['LAW']

new_ner.head()
```

```python
new_ner.isna().sum()
```

```python
new_ner.describe()
```

```python
# splitting the dataset in two groups
new_ner_h = new_ner[new_ner['memType']=='recalled'].drop('memType', axis=1)
new_ner_d = new_ner[new_ner['memType']=='imagined'].drop('memType', axis=1)

# checking assumptions of normality and homogeneity of variance: 
for col in new_ner_h.columns:
    # check normality
    _, p_val1 = shapiro(new_ner_h[col])
    _, p_val2 = shapiro(new_ner_d[col])
    if p_val1 < 0.05 or p_val2 < 0.05:
        print(f"{col}: not normally distributed")
    else:
        print(f"{col}: normally distributed")
        
    # check equal variances
    _, p_val = levene(new_ner_h[col], new_ner_d[col])
    if p_val < 0.05:
        print(f"{col}: variances are not equal")
    else:
        print(f"{col}: variances are equal")
```

```python
# Computing non parametric indipendent sample t-tests:

t_stats = []
p_values = []

for col in new_ner_h.columns:
    t_stat, p_val = mannwhitneyu(new_ner_h[col], new_ner_d[col])
    t_stats.append(t_stat)
    p_values.append(p_val)

# perform multiple comparisons correction using Bonferroni method
null_h_rej, corrected_p_values, _, _ = multipletests(p_values, alpha = 0.05, method='holm')

# save results in a dataframe
t_tests_df = pd.DataFrame({"feature": new_ner_h.columns, "t_stat": t_stats, "p_value": p_values, "corrected_p_value": corrected_p_values, "null_h_rejected": null_h_rej})
t_tests_df["significance"] = ['*' if p == True else '-' for p in t_tests_df['null_h_rejected']]

# print all comparisons
#t_tests_df.sort_values("null_h_rejected", ascending=False, inplace=True)
t_tests_df
```

```python
columns = t_tests_df.loc[t_tests_df['null_h_rejected'] == True, 'feature'].tolist()
columns.remove('word_counts')

# Calculate the average and standard deviation for each column in both datasets
new_ner_h_mean = new_ner_h[columns].mean()
new_ner_h_std = new_ner_h[columns].std()
new_ner_d_mean = new_ner_d[columns].mean()
new_ner_d_std = new_ner_d[columns].std()

# Create a bar plot
fig, ax = plt.subplots(figsize=(12, 6))
ind = np.arange(len(columns)) 
width = 0.35 

rects1 = ax.bar(ind - width/2, new_ner_h_mean, width, yerr=new_ner_h_std,label='Recalled')
rects2 = ax.bar(ind + width/2, new_ner_d_mean, width, yerr=new_ner_d_std,label='Imagined')

ax.set_xlabel('Features')
ax.set_ylabel('Values')
ax.set_title('Mean and Std of Each Entity for Imagined and Recalled Stories')
ax.set_xticks(ind)
ax.set_xticklabels(columns, rotation=45, ha="right")
ax.legend()

plt.show()
```
